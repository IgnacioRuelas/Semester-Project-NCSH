```python
import pandas as pd
import streamlit as st

df = pd.read_csv('/content/NCHS_Data.csv')

# Filter for applications that are 'Ready for Review'
ready_df = df[df["Request Status"] == "Approved"]

# Sidebar filter for signed status
signed_filter = st.sidebar.selectbox(
    "Filter by Application Signed ?",
    options=["Yes", "No", "Missing", "N/A"]
)

st.title("Applications Ready for Review")
st.write(f"Showing {len(ready_df)} applications")

st.dataframe(ready_df)
```
